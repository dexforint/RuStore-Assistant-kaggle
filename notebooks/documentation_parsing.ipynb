```python
import requests
from bs4 import BeautifulSoup
import pickle

DOC_FOLDER = "./documentation"
```

```python
import time
from copy import copy
docs = []

links_to_explore = set()

def parse_url(current_url, depth=0):
    """Рекурсивный обход списка"""
    
    global docs, links_to_explore
    
    urls = [doc['url'] for doc in docs]

    if current_url in urls:
        if depth == 0:
            links_to_explore_ = copy(links_to_explore)
            for link in links_to_explore_:
                parse_url(link, 1)
        
        return
    
    response = requests.get(current_url) # !!!
    content = response.content.decode('utf-8')
    soup = BeautifulSoup(content, 'html.parser')

    doc = soup.find('div', {'class': 'theme-doc-markdown'})
    if not doc:
        doc = soup.find('header')

    assert doc, doc

    breadcrumbs = soup.find_all('a', {'class': 'menu__link--active'})
    breadcrumbs = [active_link.text.strip() for active_link in breadcrumbs]
    
    breadcrumbs = [text.replace("\x00",  "") for text in breadcrumbs]
    

    docs.append({
        "breadcrumbs": breadcrumbs,
        "url": current_url,
        "html": str(doc)
    })
    links_to_explore.discard(current_url)
    time.sleep(1)
    print(len(docs))

    links = soup.find_all('a', {'class': 'menu__link'})
    urls = []
    for a in links:
        url = a.get('href')
        url = f"https://www.rustore.ru{url}"
        
        if url.endswith("/"):
            url = url[:-1]

        links_to_explore.add(url)
        urls.append(url)
    
    for url in urls:
        parse_url(url, 1)
```
